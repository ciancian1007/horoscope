```python
#threads
import time
import undetected_chromedriver as uc
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = uc.Chrome()
driver.get("https://www.threads.net/@twtd003?hl=zh-tw")
driver.execute_script("window.scrollTo(0, 300);")
time.sleep(3)

posts_list = driver.find_elements(By.CLASS_NAME,'x78zum5.xdt5ytf')  #先找到存放文章的標籤

for post in posts_list:
    elements = post.find_elements("css selector","div.x1a6qonq, picture, time")   #照出文章內需要的標籤
    article = []
    for element in elements:  #將每個標籤內容按順序印出
        if element.tag_name == "time":
            article.append("=====\n發文時間:"+element.get_attribute("title"))
        elif element.tag_name == "div":
            article.append(element.text)
        elif element.tag_name == "picture":
            article.append(element.find_element(By.TAG_NAME,"img").get_attribute("src"))
    article = "".join(article).split("======") #將文章進行切割
    for i in article:
        print(i)

driver.close()
```

```python
#dcard爬內文
import time
import undetected_chromedriver as uc
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = uc.Chrome()
driver.get("https://www.dcard.tw/f/horoscopes/p/257077404")
time.sleep(3)

article = driver.find_element(By.TAG_NAME, "article")
post_time = article.find_element(By.XPATH, ".//time").get_attribute("title")

elements = article.find_elements(By.XPATH,".//h1 | .//span | .//picture | .//time | .//a[@target='_blank']")
for element in elements:
    if element.tag_name == "h1":
        print("Title:", element.text, end="\n")
    elif element.tag_name == "time":
        print("post time:",element.get_attribute("title"), end="\n")
    elif element.tag_name == "span":
        print(element.text, end="\n")
    elif element.tag_name == "picture":
        print(element.find_element(By.TAG_NAME,"img").get_attribute("src"), end="\n")
    elif element.tag_name == "a":
        print(element.get_attribute("href"), end="\n")
driver.close()
```
